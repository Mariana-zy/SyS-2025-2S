# Serie de Fourier (exponencial, trigonom√©trica y compacta)

## Semejanzas

* Las tres sirven para representar una se√±al peri√≥dica como la suma de muchas ondas m√°s simples de distintas frecuencias.

* Todas permiten ver de qu√© frecuencias est√° compuesta una se√±al y con qu√© amplitud y fase aparece cada una.

* Son equivalentes: dicen lo mismo, solo que con diferente forma matem√°tica.

## Diferencias

* **Forma trigonom√©trica:**

Usa funciones reales: senos y cosenos.

$$ x(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} [ a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t) ] $$



Es la m√°s f√°cil de entender visualmente, porque muestra claramente cu√°nto ‚Äúcoseno‚Äù y cu√°nto ‚Äúseno‚Äù hay en la se√±al.

Los coeficientes ùëéùëõ y ùëèùëõ indican la amplitud de cada componente.

* **Forma exponencial o compleja:**

Usa exponentes complejos en lugar de senos y cosenos:

$$ x(t) = \sum_{k=-\infty}^{\infty} C_k \, e^{j k \omega_0 t} $$

Es m√°s f√°cil de manejar en c√°lculos, sobre todo en ingenier√≠a.

Los coeficientes ùê∂ùëò son n√∫meros complejos que guardan la misma informaci√≥n que ùëéùëõ y ùëèùëõ, pero combinada en una sola expresi√≥n.

* **Forma compacta o fasorial:**

Esta forma muestra la se√±al como una combinaci√≥n de muchas ondas cosenoidales que ‚Äúencajan‚Äù entre s√≠.

$$ x(t) = C_{0} + \sum_{n=1}^{\infty} C_{n} \cos (n\omega_{0}t + \theta_{n}) $$

La primera (ùëõ=1) es la frecuencia fundamental, y las dem√°s (ùëõ>1) son los arm√≥nicos, que dan forma y detalle a la se√±al.

Permite ver claramente la amplitud y fase de cada componente, lo que la hace muy √∫til para interpretar c√≥mo se construye una se√±al peri√≥dica a partir de sus arm√≥nicos.

## Observaci√≥n:

Las tres formas representan el mismo fen√≥meno: c√≥mo una se√±al peri√≥dica puede formarse a partir de ondas m√°s simples. La forma trigonom√©trica es la m√°s intuitiva, mientras que la exponencial y la compacta son m√°s pr√°cticas para el an√°lisis y los c√°lculos.


# Transformada de Fourier Transformada de Fourier en Tiempo Discreto y Transformada Discreta de Fourier

## Semejanzas

* Las tres herramientas permiten analizar una se√±al en el dominio de la frecuencia, mostrando qu√© frecuencias componen la se√±al.
* Todas buscan relacionar tiempo y frecuencia, mostrando la misma informaci√≥n desde dos perspectivas diferentes.  
* Son matem√°ticamente relacionadas: la DTFT y la DFT son versiones discretas o muestreadas de la Transformada de Fourier continua.  
* En todos los casos, existe una transformada directa (para pasar al dominio de la frecuencia) y una transformada inversa (para reconstruir la se√±al original).  

## Diferencias

* **Transformada de Fourier (TF):**

Se aplica a se√±ales en tiempo continuo y generalmente no peri√≥dicas.  
El resultado es un espectro continuo en frecuencia.  

$$ X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} \, dt $$

y su inversa:

$$ x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\omega) e^{j\omega t} \, d\omega $$

Permite ver c√≥mo se distribuye la energ√≠a de una se√±al continua en todas las frecuencias.

Por ejemplo, una se√±al de audio real se analiza con la Transformada de Fourier continua.

* **Transformada de Fourier en Tiempo Discreto (DTFT):**

Se aplica a se√±ales discretas (muestreadas en el tiempo), pero que pueden ser infinitas en duraci√≥n.  
El resultado es un espectro continuo en frecuencia, aunque peri√≥dico con periodo \(2\pi\).  

$$ X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n} $$

y su inversa:

$$ x[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\omega}) e^{j\omega n} \, d\omega $$

La periodicidad del espectro aparece porque muestrear en el tiempo produce repeticiones del espectro en frecuencia.  
Se usa para analizar sistemas digitales con se√±ales discretas.

* **Transformada Discreta de Fourier (DFT):**

Se aplica a se√±ales discretas y de duraci√≥n finita (N muestras).  
El resultado es un espectro discreto y peri√≥dico, con N puntos en frecuencia.  

$$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j\frac{2\pi}{N}kn}, \quad k = 0, 1, \dots, N-1 $$

y su inversa:

$$ x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j\frac{2\pi}{N}kn} $$

La DFT es la versi√≥n num√©rica y pr√°ctica de la transformada, ya que se puede calcular en computador.  
Cuando se usa el algoritmo FFT (Fast Fourier Transform), el c√°lculo se vuelve mucho m√°s eficiente.

## Relaci√≥n entre se√±al y espectro

- Se√±al continua ‚Üí Espectro continuo: Transformada de Fourier (TF).  
- Se√±al discreta ‚Üí Espectro peri√≥dico (continuo en frecuencia): DTFT.  
- Se√±al discreta y finita ‚Üí Espectro discreto y peri√≥dico: DFT.  

Al discretizar la se√±al, el espectro se vuelve peri√≥dico.  
Y al hacer la se√±al finita (tomar solo N muestras), el espectro pasa a tener un n√∫mero finito de componentes.

## Observaci√≥n

Las tres transformadas describen el mismo principio de an√°lisis de frecuencia, pero en diferentes contextos:

- La TF trabaja con se√±ales anal√≥gicas (tiempo continuo).  
- La DTFT analiza se√±ales digitales infinitas.  
- La DFT es la versi√≥n computacional, usada en procesadores y software (por ejemplo, con el algoritmo FFT).  

Todas buscan descomponer una se√±al en sus frecuencias b√°sicas.  
La diferencia est√° en si la se√±al y el espectro son continuos o discretos, lo que depende de si se trabaja en el mundo f√≠sico (anal√≥gico) o digital (muestreado).

#  Transformada R√°pida de Fourier (FFT): Qu√© es, por qu√© sirve y c√≥mo funciona

## Semejanzas (FFT vs DFT)

* Tanto la DFT (Transformada Discreta de Fourier) como la FFT (Transformada r√°pida de Fourier) proporcionan los mismos valores de frecuencia para una secuencia de (N) muestras: ambas calculan los (X[k]) que representan el contenido en frecuencia de (x[n]).  
* Ambas usan la misma f√≥rmula matem√°tica b√°sica:
  $$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j\frac{2\pi}{N}kn}, \quad k = 0,\dots,N-1 $$
* La FFT no es otra transformada distinta: es un algoritmo para calcular la DFT m√°s r√°pido.

## Diferencias

* **DFT (c√°lculo directo):**

  Calcula la suma directamente siguiendo la f√≥rmula anterior.

  Requiere aproximadamente (N^2) multiplicaciones complejas.

  Es costosa para \(N\) grandes.

* **FFT (Fast Fourier Transform):**
  
  Es un m√©todo/algoritmo para calcular exactamente la misma DFT pero con muchas menos operaciones.

  Reduce el costo a Nlog(N) operaciones.
  
  Para (N) grande, la reducci√≥n de c√≥mputo es enorme y hace posible usar la DFT en tiempo real y en aplicaciones pr√°cticas.

## ¬øEn qu√© consiste la idea b√°sica de la FFT?

1. **Descomposici√≥n recursiva:** Dividir la DFT de tama√±o (N) en DFTs m√°s peque√±as, por ejemplo, dos DFTs de tama√±o (N/2) aprovechando que los √≠ndices pares e impares se separan f√°cilmente.  

2. **Reusar c√°lculos:** Las DFT peque√±as se combinan con multiplicaciones por factores llamados twiddle factors, son factores de rotaci√≥n compleja:
$$ W_N^k = e^{-j \, \frac{2\pi k}{N}} $$

3. **Recursividad:** Se repite la divisi√≥n hasta llegar a DFTs de tama√±o 1.

4. **Combinar con butterflies:** La combinaci√≥n se hace mediante operaciones simples llamadas *butterfly* (cada butterfly mezcla dos valores mediante suma/resta y una multiplicaci√≥n por un twiddle).

**Resultado:** En lugar de calcular (N^2) productos, se hacen muchas menos operaciones porque se reutiliza la informaci√≥n de las DFTs peque√±as.

## Observaci√≥n

La FFT es el algoritmo que hizo pr√°ctica la DFT en el mundo real. Gracias a la reducci√≥n de O(N^2) a O(N\log(N)), es posible procesar audio, imagen, comunicaciones y se√±ales biom√©dicas en tiempo real y en dispositivos con recursos limitados. En la pr√°ctica, siempre que se necesite la DFT de una secuencia (especialmente para N grande), casi siempre se usar√° una FFT optimizada.
