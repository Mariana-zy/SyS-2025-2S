```python
#   BODE DEL CIRCUITO RLC EN SERIE
#   R = 1k ohm, L = 180 mH, C = 120 uF

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parmetros del circuito
R = 1e3          # ohm
L = 180e-3       # H
C = 120e-6       # F

# Coeficientes de la funcin de transferencia H(s)
# H(s) = 1 / (LC s^2 + RC s + 1)
num = [1]
den = [L*C, R*C, 1]

# Crear sistema
sistema = signal.TransferFunction(num, den)

# Crear rango de frecuencias logartmico
w = np.logspace(0, 5, 3000)  # desde 1 rad/s hasta 1e5 rad/s

# Obtener bode
w, mag, fase = signal.bode(sistema, w=w)

# GRFICA DE MAGNITUD
plt.figure(figsize=(10,6))
plt.semilogx(w, mag)
plt.title("Diagrama de Bode - Magnitud")
plt.ylabel("Magnitud (dB)")
plt.xlabel("Frecuencia (rad/s)")
plt.grid(True, which="both")
plt.show()

# GRFICA DE FASE
plt.figure(figsize=(10,6))
plt.semilogx(w, fase)
plt.title("Diagrama de Bode - Fase")
plt.ylabel("Fase ()")
plt.xlabel("Frecuencia (rad/s)")
plt.grid(True, which="both")
plt.show()
```
